```python
# Standard inputs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import pickle
%matplotlib inline
```

```python
df = pd.read_csv("clean_dataset.csv")
df.head()
```

```python
# Split into X/y and drop un-used column "CreditScore"
X = df.drop(columns=["Approved", "CreditScore"], errors="ignore")
y = df["Approved"]

# Split into training and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Values must be converted to numerical values
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Define categorical features
cf = ["Industry", 
      "Ethnicity", 
      "Citizen"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot",
                                  one_hot, 
                                  cf)], remainder="passthrough")
transformed_X = transformer.fit_transform(X)
#transformed_X
```

```python
# Refit model
np.random.seed(45)
X_train, X_test, y_train, y_test = train_test_split(transformed_X, y, test_size=0.20)

from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
# Accuracy
model.score(X_test, y_test)
```

```python
with open("model.pkl", "wb") as f:
    pickle.dump(model, f)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier

np.random.seed(52)

X = df.drop(columns=["Approved", "CreditScore"], errors="ignore")
y = df["Approved"]

X = pd.get_dummies(X)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15)

model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
cross_val_score(model, X, y)
```

```python
np.random.seed(43)

single = model.score(X_test, y_test)

cross_val = np.mean(cross_val_score(model, X, y))

single, cross_val
```

```python
from sklearn.metrics import confusion_matrix

y_predict = model.predict(X_test)

confusion_matrix(y_test, y_predict)
```

```python
pd.crosstab(y_test, y_predict, rownames=["actual"], colnames=["predicted"])
```
