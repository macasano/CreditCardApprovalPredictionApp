```python
from tkinter import *
from tkinter import ttk
import numpy as np
import pandas as pd
import pickle
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

with open("model.pkl", "rb") as f:
    model = pickle.load(f)

panel = Tk()
panel.title("C964 Capstone")
panel.geometry("800x700")
panel.resizable(False,True)
panel.bell(displayof=0)

Label(panel,text = "Credit Card Application",font = "arial 20", fg = "blue").pack(pady=30)
Label(text = "-----------------------------------------------------------------------------------------------------------------------------------", font = "arial 11").place(x=90,y=110)
Label(text = "What is your gender?", font = "arial 12").place(x=100,y=95)
Label(text = "-----------------------------------------------------------------------------------------------------------------------------------", font = "arial 11").place(x=90,y=145)
Label(text = "How old are you? (years)", font = "arial 12").place(x=100,y=130)
Label(text = "-----------------------------------------------------------------------------------------------------------------------------------", font = "arial 11").place(x=90,y=175)
Label(text = "How much debt do you have?", font = "arial 12").place(x=100,y=160)
Label(text = "-----------------------------------------------------------------------------------------------------------------------------------", font = "arial 11").place(x=90,y=205)
Label(text = "Are you married?", font = "arial 12").place(x=100,y=190)
Label(text = "-----------------------------------------------------------------------------------------------------------------------------------", font = "arial 11").place(x=90,y=235)
Label(text = "Are you a bank customer", font = "arial 12").place(x=100,y=220)
Label(text = "-----------------------------------------------------------------------------------------------------------------------------------", font = "arial 11").place(x=90,y=265)
Label(text = "What industry do you work in?", font = "arial 12").place(x=100,y=250)
Label(text = "-----------------------------------------------------------------------------------------------------------------------------------", font = "arial 11").place(x=90,y=295)
Label(text = "Ethinicity", font = "arial 12").place(x=100,y=280)
Label(text = "-----------------------------------------------------------------------------------------------------------------------------------", font = "arial 11").place(x=90,y=325)
Label(text = "How many years have you been employed?", font = "arial 12").place(x=100,y=310)
Label(text = "-----------------------------------------------------------------------------------------------------------------------------------", font = "arial 11").place(x=90,y=355)
Label(text = "Have you defaulted on a payment?", font = "arial 12").place(x=100,y=340)
Label(text = "-----------------------------------------------------------------------------------------------------------------------------------", font = "arial 11").place(x=90,y=385)
Label(text = "Are you employed?", font = "arial 12").place(x=100,y=370)
Label(text = "-----------------------------------------------------------------------------------------------------------------------------------", font = "arial 11").place(x=90,y=415)
Label(text = "Do you have a driver's license?", font = "arial 12").place(x=100,y=400)
Label(text = "-----------------------------------------------------------------------------------------------------------------------------------", font = "arial 11").place(x=90,y=445)
Label(text = "Citizenship", font = "arial 12").place(x=100,y=430)
Label(text = "-----------------------------------------------------------------------------------------------------------------------------------", font = "arial 11").place(x=90,y=475)
Label(text = "What is your zipcode? (5-digits)", font = "arial 12").place(x=100,y=460)
Label(text = "-----------------------------------------------------------------------------------------------------------------------------------", font = "arial 11").place(x=90,y=505)
Label(text = "What is your monly income?", font = "arial 12").place(x=100,y=490)

Label(text = "Prediction", font = 50).place(x=100,y=530)

# Dropdown Menus
gender = ["male","female"]
cb_gender = ttk.Combobox(panel, values=gender)
cb_gender.pack()
cb_gender.place(x=600,y=100)

married = ["yes", "no"]
cb_married = ttk.Combobox(panel, values=married)
cb_married.pack()
cb_married.place(x=600,y=190)

bank_cust = ["yes", "no"]
cb_bnkcust = ttk.Combobox(panel, values=married)
cb_bnkcust.pack()
cb_bnkcust.place(x=600,y=220)

industry = ["Industrials", "Materials","CommunicationServices","Transport","InformationTechnology","Financials","Energy","Real Estate","Utilities","ConsumerDiscretionary","Education","ConsumerStaples","Healthcare","Research"]
cb_industry = ttk.Combobox(panel, values=industry)
cb_industry.pack()
cb_industry.place(x=600,y=250)

ethnicity = ["White", "Black","Asian","Latino","Other"]
cb_ethnicity = ttk.Combobox(panel, values=ethnicity)
cb_ethnicity.pack()
cb_ethnicity.place(x=600,y=280)

defaulted = ["yes", "no"]
cb_default = ttk.Combobox(panel, values=defaulted)
cb_default.pack()
cb_default.place(x=600,y=340)

employed = ["yes", "no"]
cb_employed = ttk.Combobox(panel, values=employed)
cb_employed.pack()
cb_employed.place(x=600,y=370)

license = ["yes", "no"]
cb_license = ttk.Combobox(panel, values=license)
cb_license.pack()
cb_license.place(x=600,y=400)

citizen = ["ByBirth", "ByOtherMeans","Temporary"]
cb_citizen = ttk.Combobox(panel, values=citizen)
cb_citizen.pack()
cb_citizen.place(x=600,y=430)


# Entry Values
ageValue = StringVar()
appAge = Entry(panel, textvariable=ageValue, width=20,bd=1, font= "arial 10")
appAge.place(x=600, y=130)

debtValue = StringVar()
appDebt = Entry(panel, textvariable=debtValue, width=20,bd=1, font= "arial 10")
appDebt.place(x=600, y=160)

yrsEmpValue = StringVar()
appYrsEmp = Entry(panel, textvariable=yrsEmpValue, width=20,bd=1, font= "arial 10")
appYrsEmp.place(x=600, y=310)

zipcodeValue = StringVar()
appzipcode = Entry(panel, textvariable=zipcodeValue, width=20,bd=1, font= "arial 10")
appzipcode.place(x=600, y=460)

incomeValue = StringVar()
appIncome = Entry(panel, textvariable=incomeValue, width=20,bd=1, font= "arial 10")
appIncome.place(x=600, y=490)

def to_binary(x):
    if x == "yes" or x == "female":
        return 1
    else:
        return 0

def on_button_click():
    try:
        required_data = [
            cb_gender.get(), ageValue.get(), debtValue.get(), cb_married.get(), cb_bnkcust.get(), cb_industry.get(),
            cb_ethnicity.get(), yrsEmpValue.get(), cb_default.get(), cb_employed.get(), cb_license.get(), cb_citizen.get(), 
            zipcodeValue.get(), incomeValue.get()
        ]

        age = round(float(ageValue.get()), 2)
        yrsEmp = round(float(yrsEmpValue.get()), 2)
        
        if len(zipcodeValue.get()) < 5:
            Label(text="Please enter a 5-digit zip code.", font="arial 10").place(x=400, y=460)
            return

        if age <= 17:
            Label(text="You must be at least 18 years old to apply.").place(x=400, y=130)
            
        if age < yrsEmp:
            Label(text="Age cannot be less than years employed.").place(x=400, y=130)
        
        if any(f == '' for f in required_data):
            Label(text="Please fill in all fields on the application before submitting.", font="arial 10").place(x=250, y=650)
            return
            
        appData = {
            "Gender": [to_binary(cb_gender.get())],
            "Age": [round(float(ageValue.get()), 0)],
            "Debt": [round(float(debtValue.get()), 2)],
            "Married": [to_binary(cb_married.get())],
            "BankCustomer": [to_binary(cb_bnkcust.get())],
            "Industry": [cb_industry.get()],
            "Ethnicity": [cb_ethnicity.get()],
            "YearsEmployed": [round(float(yrsEmpValue.get()), 0)],
            "PriorDefault": [to_binary(cb_default.get())],
            "Employed": [to_binary(cb_employed.get())],
            "DriversLicense": [to_binary(cb_license.get())],
            "Citizen": [cb_citizen.get()],
            "ZipCode": [round(float(zipcodeValue.get()), 0)],
            "Income": [round(float(incomeValue.get()), 2)]
        }

        with open("model.pkl", "rb") as f:
            model = pickle.load(f)

        df = pd.read_csv("clean_dataset.csv")

        features = df.drop(columns=["Approved", "CreditScore"]).columns.tolist()

        columns = ["Industry", "Ethnicity", "Citizen"]
        
        transformer = ColumnTransformer(
            transformers=[("one_hot", OneHotEncoder(), columns)], 
            remainder="passthrough")

        transformer.fit(df[features])

        appDatadf = pd.DataFrame(appData)

        transformed_data = transformer.transform(appDatadf[features])

        prediction = model.predict(transformed_data)
        print(f"Prediction: {prediction}")

        # Prediction Status
        result_status = "Approved" if prediction[0] == 1 else "Denied"
        pred_status = Label(panel, text=f"{result_status}")
        pred_status.config(text=f"{result_status}")
        pred_status.place(x=250, y=540)

    except Exception as e:
        print("Error:", e)

button = Button(panel, 
                text="Submit", 
                command=on_button_click)
button.place(x=400,y=620)
        
panel.mainloop()
```
